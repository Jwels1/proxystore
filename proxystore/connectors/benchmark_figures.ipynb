```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fig, ax = plt.subplots()
```

```python
ax.set_xscale('log')
```

```python
amount_encrypted = {1: .2117, 10: .241, 100: .510, 1000: 3.064, 10000: 29.351}
amount_unencrypted = {1: .187, 10: .212, 100: .476, 1000: 2.92, 10000: 25.143}
```

```python
ax.plot(amount_encrypted.keys(), amount_encrypted.values(), marker = "o", color = "b", label = "encrypted")
ax.plot(amount_unencrypted.keys(), amount_unencrypted.values(), marker = "*", color = "r", label = "unencrypted")
ax.legend()
ax.set_ylabel("Time (seconds)")
ax.set_xlabel("Number of Proxies (8 bytes)")
```
